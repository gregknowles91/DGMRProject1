```python
# Dependencies
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np

# Import the API key
from config import api_key
```

```python
# Build the endpoint URL

begin_date = "01-2000"
end_date = "12-2023"
time_frame = f'?from={begin_date}&to={end_date}'
crime = "robbery" # homicide // rape // robbery
level = "national"
state = "TX"
print(time_frame)
# elif crime == "Property":
#     crime == "supplemental"








base_url = f'https://api.usa.gov/crime/fbi/cde/summarized/national/{crime}{time_frame}{api_key}'
# base_url = 'https://api.usa.gov/crime/fbi/cde/summarized/national/homicide?from=01-2000&to=12-2023'
# base_url = 'https://api.usa.gov/crime/fbi/cde/summarized/national/rape?from=01-2000&to=12-2023'
# base_url = ''
```

```python
# Run a request to endpoint and convert result to json
crime_data = requests.get(base_url).json()
df = pd.DataFrame(crime_data)
# df.to_csv("Homicide_data.csv", index=False)
# df = pd.read_json(crime_data)
pprint(crime_data)


```

```python
months = []
total_crime = []
total_clearence = []
crime_per_capita = []
clear_per_capita = []
total_pop = []
participated_pop = []
pop_coverage = []
perc_pros = []

# for index in df["actuals"]["United States"]:
#     months.append(index)
#     total_homicides.append(df["actuals"]["United States"][index])
# for index in df["rates"]["United States"]:
#     homicides_per_capita.append(df["rates"]["United States"][index])
# for index in df["populations"]["participated_population"]["United States"]:
#     participated_pop.append(df["populations"]["participated_population"]["United States"][index])   
for index in df["offenses"]["actuals"]["United States"]:
    months.append(index)
    total_crime.append(df["offenses"]["actuals"]["United States"][index])
for index in df["offenses"]["actuals"]["United States Clearances"]:
    total_clearence.append(df["offenses"]["actuals"]["United States Clearances"][index])
for index in df["offenses"]["rates"]["United States"]:
    crime_per_capita.append(df["offenses"]["rates"]["United States"][index])
for index in df["offenses"]["rates"]["United States Clearances"]:
    clear_per_capita.append(df["offenses"]["rates"]["United States Clearances"][index])
for index in df["populations"]["population"]["United States"]:
    total_pop.append(df["populations"]["population"]["United States"][index])
for index in df["populations"]["participated_population"]["United States"]:
    participated_pop.append(df["populations"]["participated_population"]["United States"][index])
for index in df["tooltips"]["Percent of Population Coverage"]["United States"]:
    pop_coverage.append(df["tooltips"]["Percent of Population Coverage"]["United States"][index])

df_testing = pd.DataFrame(total_crime, index=months, columns=[f"Total {crime}"])
df_testing["Clearances"] = total_clearence
df_testing["Crime(Per 100k)"] = crime_per_capita
df_testing["Prosecutions(Per 100k)"] = clear_per_capita
df_testing["Pros/Perc"] = (df_testing["Clearances"] / df_testing[f"Total {crime}"]).round(2)
df_testing["Total Pop"] = total_pop
df_testing["Partic Pop"] = participated_pop
df_testing["Pop Coverage"] = pop_coverage
df_testing
```

```python
df_testing.index = pd.to_datetime(df_testing.index, format='%m-%Y', errors='coerce')
df_sorted = df_testing.sort_index()
df_sorted
```
